```python
#Rectangle 
class Rectangle:
    def __init__(self, length, width):
        self.width = width
        self.length = length
        
    def area(self):
        result = self.width * self.length
        return result
    
    def perimeter(self):
        result = 2 * (self.width * self.length)
        return result
    
    def display(self):
        print(f"Rong: {self.width}, dai: {self.length}, chu vi: {self.perimeter():.2f}, dien tich: {self.area():.2f}")
```

```python
menu_options = {
    1: 'Doc du lieu tu file input.db',
    2: 'Them moi hinh chu nhat',
    3: 'Hien thi danh sach hinh chu nhat',
    4: 'Tinh tong dien tich cac hanh chu nhat ',
    5: 'Hien thi cac hinh chu nhat co chu vi nho nhat',
    6: 'Luu danh sach hinh chu nhat',
    'Other': 'Thoat chuong trinh'
}

# show key and value of dictionary
def print_menu():
    for key in menu_options.keys():
        print(key, '--', menu_options[key]) 
        
#khai bao bien luu tru hcn

while(True):
    print_menu()
    try: 
        userChoice = int(input("Nhap tuy chon: "))
    except:
        print('Nhap sai dinh dang, hay nhap lai,...')
        continue
    if userChoice == 1:
        fr = open('input.txt', 'r', encoding='utf-8')
        dsHCN = []
        for line in fr:
            stripLine = line.strip("\n")
            ds = stripLine.split(',')
            cd = float(ds[0])
            cr = float(ds[1])
            hcn = Rectangle(cd,cr)
            dsHCN.append(hcn)
    elif userChoice == 2:
        cd = float(input("Nhap chieu dai: "))
        cr = float(input("Nhap chieu rong: "))
        
        hcn = Rectangle(cd, cr)
        dsHCN.append(hcn)
    elif userChoice == 3:
        for item in dsHCN:
            item.display()
    elif userChoice == 4:
        sum_area = 0
        for item in dsHCN:
            sum_area = sum_area + item.area()
        
        print(f"Tong dien tinh: {sum_area}")
    elif userChoice == 5:
        if dsHCN.count == 0:
            print('Danh sach rong')
        else:
            chu_vi = dsHCN[0].perimeter()
            for item in dsHCN:
                if chu_vi > item.perimeter():
                    chu_vi = item.perimeter()
            for item in dsHCN:
                if item.perimeter() == chu_vi:
                    item.display()
    elif userChoice == 6:
        fw = open('outputdemo4.txt', 'w', encoding = 'utf-8')
        for item in dsHCN:
            fw.write(f"{item.length} - {item.width} - {item.perimeter()} - {item.area()}\n")
        fw.close
    else:
        print("Thoat chuong trinh BYE BYE")
        break
    
    
    # if create list save Object -> The list can use method of Object
```

```python
class Dog:
    dog_count = 0
    def __init__(self, name, size, age, color):
        self.name = name 
        self.size = size
        self.age = age
        self.color = color
        Dog.dog_count = Dog.dog_count + 1
    
    def go(self):
        print("I am going...")
        
    def stay(self, place):
        print("I am staying at {}".format(place))
        
    def lie(self, place):
        print("I am lying at {}".format(place))
        
    def bark(self):
        print("Whoop...")
        
    def eat(self, food):
        print("I am eating {}".format(food))
        
    def create_dog(self, name, size, age, color):
        Dog.dog_count = Dog.dog_count + 1
        return Dog(name, size, age, color)
    
    def report():
        print("Number of dogs: {}".format(Dog.dog_count))

bull = Dog("Bull", "lagre", 2, "yellow")
bull.stay("garden")
bull.bark()

lucky = Dog("lucky", "lagre", 8, "black")

Dog.report()
```
